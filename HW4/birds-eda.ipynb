```python
import os
directories = ['../input/csc4851-homework4/birds_400/test',
                                '../input/csc4851-homework4/birds_400/train',
                                '../input//csc4851-homework4/birds_400/valid']

for dir in directories:
    label = []
    path = []
    for dirname, _,filenames in os.walk(dir):
        for filename in filenames:
            label.append(os.path.split(dirname)[1])
            path.append(os.path.join(dirname,filename))
    if dir == directories[0]:
        df_test = pd.DataFrame(columns=['path','label'])
        df_test['path']=path
        df_test['label']=label
    elif dir == directories[1]:
        df_train = pd.DataFrame(columns=['path','label'])
        df_train['path']=path
        df_train['label']=label        
    elif dir == directories[2]:
        df_valid = pd.DataFrame(columns=['path','label'])
        df_valid['path']=path
        df_valid['label']=label
```

```python
def img(path, list_of_filename, size = (64, 64)):
    for file in list_of_filename:
        fpath = path + file
        current_image = image.load_img(fpath, target_size = size, 
                                       color_mode = 'grayscale')
        img1 = image.img_to_array(current_image)
        img1 = [img1.ravel()]
        try:
            full = np.concatenate((full, img1))
        except UnboundLocalError: 
            full = img1
    return full
```

```python
train_images = {}
num_classes = 5
train_images_list = {}
classes = [x[0].split('/')[-1] for x in os.walk(directory+"birds_400/train/")]

for class1 in classes[:num_classes]:
    list_of_names = list(os.walk(directory+"birds_400/train/"+class1))
    train_images_list[class1] = list_of_names[0][2]
for class1 in classes[:num_classes]:
    x = img(f'{directory}birds_400/train/{class1}/', train_images_list[class1])
    train_images[class1] = x
```
